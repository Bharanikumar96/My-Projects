```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
```

```python
data = pd.read_csv("D:\Project\Machine learning\Desion Tree\heart.csv")
data
```

```python
data.isna().sum()
```

```python
numeric = data.select_dtypes(include=['float64','int64'])
numeric = numeric.fillna(numeric.mean())
```

```python
scaler = StandardScaler()
scaled_data = scaler.fit_transform(numeric)
```

```python
n_components = 2
pca = PCA(n_components = n_components)
pca_data = pca.fit_transform(scaled_data)
```

```python
print("Explained variance ratio by each component: ")
print(pca.explained_variance_ratio_)
```

```python
plt.figure(figsize=(8,6))
plt.scatter(pca_data[:,0],pca_data[:,1],c='blue',edgecolor='k')
plt.xlabel('List1')
plt.ylabel('List2')
plt.title('PCA Scatter Plot')
plt.show()
```

```python
pca_transformed_df = pd.DataFrame(pca_data,columns=[f'PC{i+1}'
                                                  for i in range(n_components)])
pca_transformed_df.to_csv('pca_tranformed_data.csv',index=False)
print("PCA-transformed data saved to 'pca_new_data.csv'")
```
