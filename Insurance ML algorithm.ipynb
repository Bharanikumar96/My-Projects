```python
sns.kdeplot(insur["age"])
plt.show()
```

```python
for i in insur.columns:
    if insur[i].dtypes!=object:
        sns.distplot(insur[i])
        plt.title(i)
        plt.show()
```

```python
for i in insur.columns:
    if insur[i].dtypes!=object:
        q1 = insur[i].quantile(0.25)
        q3 = insur[i].quantile(0.75)
        iqr = q3-q1
        uf = q3+1.5*iqr
        If = q1-1.5*iqr
        insur = insur[(insur[i]>=If)&(insur[i]<=uf)]
```

```python
for i in insur.columns:
    if insur[i].dtypes!=object:
        sns.boxplot(insur[i])
        plt.title(i)
        plt.show()
```

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
columnlist=[]
for i in insur.columns:
    if((insur[i].dtypes!=object)&(i!="charges")):
        columnlist.append(i)
print(columnlist)
```

```python
x = insur[columnlist]
print(x)
```

```python
vif_data = pd.DataFrame()
vif_data["feature"] = x.columns
vif_data["vif"] = [variance_inflation_factor(x.values,i)
                   for i in range(len(x.columns))]
print(vif_data)
```

```python
x = insur.loc[:,["age","bmi","children"]]
y = insur["charges"]
print(x)
print(y)
```

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest = train_test_split(x,y,test_size=0.2)
print(xtrain.shape)
print(xtest.shape)
print(ytrain.shape)
print(ytest.shape)
```

```python
from sklearn.linear_model import LinearRegression
l = LinearRegression()
l.fit(xtrain,ytrain)
pred = l.predict(xtest)
print(pred)
```

```python
from sklearn.metrics import r2_score
s = r2_score(ytest,pred)
print(s*100)
```

```python
err = pd.DataFrame(columns=["Actual","Original"])
err["Actual"] = ytest
err["Original"] = pred
print(err.head())
```

```python
err["diff"] = err["Actual"] - err["Original"]
print(err.head())
```
