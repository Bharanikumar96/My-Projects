```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
```

```python
data = pd.read_csv("D:\Project\ML PROJECT\diabetes.csv")
data
```

```python
data.head()
```

```python
numeric_data = data.select_dtypes(include=['float64','int64'])
```

```python
numeric_data = numeric_data.fillna(numeric_data.mean())
```

```python
scaler = StandardScaler()
scaled_data = scaler.fit_transform(numeric_data)
```

```python
n_components = 2
pca = PCA(n_components=n_components)
pca_data = pca.fit_transform(scaled_data)
```

```python
pca.explained_variance_ratio_
```

```python
plt.figure(figsize=(8,6))
plt.scatter(pca_data[:,0],pca_data[:,1],c='blue',edgecolor='k')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('PCA Scatter Plot')
plt.show()
```

```python
pca_transformed_df = pd.DataFrame(pca_data,columns=[f'PC{i+1}'for i in range(n_components)])
pca_transformed_df.to_csv('pca_transformed_data.csv',index=False)
print("PCA-transformed data saved to 'pca_transformed_data.csv'")
```
